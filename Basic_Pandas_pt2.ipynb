```python
#Part 2 of our Pandas Lecture
import pandas as pd
```

```python
#Merging two data frames vertically, one on top of the other

#first dataframe
roster1 = {"Name" : ["Bob", "Alice", "Joe", "Mike", "Lisa", "Alan", "Eli", "Mark", "Liz", "Jane"],
          "Major": ["Accounting", "Finance", "Chemistry", "Art", "Engineering", "Biology", "Physics", "History", "Art", "Engineering"],
          "GPA": [3.5, 3.8, 3.2, 3.7, 4.0, 3.5, 3.8, 3.2, 3.1, 3.0],
          "Age": [22, 23, 31, 40, 25, 22, 33, 31, 27, 55]          
          }

#creating a pandas dataframe
roster1_df = pd.DataFrame(roster1)

#second dataframe
roster2 = {"Name" : ["Robert", "Ashley", "Jerod", "Morgan"],
          "Major": ["Anthropology", "Physics", "Computer Science", "Environmental Science"],
          "GPA": [3.5, 3.8, 3.2, 2.5],
          "Age": [19, 18, 27, 33]          
          }

#creating a pandas dataframe
roster2_df = pd.DataFrame(roster2)
```

```python
#displaying the roster1_df pandas dataframe - if you are not using a notebook write print(roster1_df)
roster1_df
```

```python
#displaying the roster2_df pandas dataframe - if you are not using a notebook write print(roster2_df)

roster2_df
```

```python
#Stacking one dataframe on top of another using the pandas .concat() method
#Put the dataframes which you wish to stack in a list and pass that list to .concat(). Set the ignore_index flag to True
#All the dataframes in the list need to be similar. That is, they should have the same column names and data types
combined_df = pd.concat([roster1_df, roster2_df] ,ignore_index= True)
```

```python
#printing the combined data frame - if you are not using a notebook write print(combined_df)

combined_df
```

```python
#You can sort a datframe by a paricular column by using the .sort_values method and by specifying the column to sort on
#e.g. dataframe.sort_values(by = 'INSERT_NAME_OF_COLUMN')
#- if you are not using a notebook write print(combined_df.sort_values(by='Age'))
combined_df.sort_values(by='Age')
```

```python
#You can also list the top N rows after performing a sort by using the list slicing technique
#For example: dataframe.sort_values(by = 'INSERT_NAME_OF_COLUMN')[start_row: end_row]
#The following example will return the top two rows sorted by the age column in descending order
#- if you are not using a notebook write print(combined_df.sort_values(by='Age', ascending= False)[0:2])

combined_df.sort_values(by='Age', ascending= False)[0:2]
```

```python
#ANOTHER WAY TO SORT AND LIST THE TOP N rows
#That is, sort the dataframe and then assign such to a new object
#e.g. new_dataframe = dataframe.sort_values(by = 'INSERT_NAME_OF_COLUMN')
df3 = combined_df.sort_values(by='Age')
```

```python
#We can find which rows in a dataframe match a particular condition
#Consider the line below as a preamble to making a filter mask
#As you can see, the rows which match the condition are labeled True and vice versa
#- if you are not using a notebook write print(combined_df['Major'] == 'Art' )

combined_df['Major'] == 'Art' 
```

```python
#Below we are assigning the conditional statement above to a variable
filter_mask = combined_df['Major'] == 'Art' 
```

```python
#Applying the filter mask created above will return the rows in the dataframe which match the filter mask's condition
#- if you are not using a notebook write print(combined_df[filter_mask])

combined_df[filter_mask]
```

```python
#The .loc method can applied to a filter mask
#Below is another way of doing the same task described above
#- if you are not using a notebook write print(combined_df.loc[filter_mask])

combined_df.loc[filter_mask]
```

```python
#Creating a filter mask which has multiple conditions - make note of the logical & operator 
# and the placements of the ') and '('
filter_mask = (combined_df['Major'] == 'Art') & (combined_df['Age'] > 30)
```

```python
#We can find which rows in a dataframe match a particular condition
#Consider the line below as a preamble to making a filter mask
#As you can see, the rows which match the condition are labeled True and vice versa
#- if you are not using a notebook write print(filter_mask)
filter_mask
```

```python
#The .loc method can applied to a filter mask
#- if you are not using a notebook write print(combined_df.loc[filter_mask])
combined_df.loc[filter_mask]
```

```python
#The negation operator '~' can be use to 'invert' a filter mask 
#That is, ~will return the rows NOT included in the original filter mask
#- if you are not using a notebook write print(combined_df.loc[~filter_mask])

combined_df.loc[~filter_mask] 
```

```python
#The .isin() method allows to filter a dataframe based on multiple values in a column.
#It takes a list of values as a parameter and is called on the dataframe
# E.g. name_of_dataframe.isn(list_of items)

#Below we are creating a filter mask which will allow us to select rows belonging to a specific majors 
majors = ['Art', 'Engineering', 'Biology']

filter_mask = combined_df['Major'].isin(majors)
```

```python
#Using the .loc method to apply the filter mask to the dataframe

#- if you are not using a notebook write print(combined_df.loc[filter_mask])

combined_df.loc[filter_mask]
```

```python
#Reminding ourselves what the original dataframe looked like
#- if you are not using a notebook write print(combined_df)
combined_df
```

```python
#How to change a record in a dataframe?
#Let's change Joe's Major from 'Chemistry' to that of 'Computer Science

#First, let's identify the index and column of the cell which we wish to change.
#In this case, it is index 2 and the 'Major' column
#Using the .loc method, we can access the the cell and change its value to 'Computer Science'
combined_df.loc[2,'Major'] = 'Computer Science'
```

```python
#printing the new dataframe to verify that the change chas been made
#- if you are not using a notebook write print(combined_df)

combined_df
```

```python
#Another filter mask example, listing all of the art majors
filter_mask = combined_df['Major'] == 'Art' 

#- if you are not using a notebook write print(combined_df[filter_mask])

combined_df[filter_mask]
```

```python
#Trying to list all the Art majors over the age of 30 without a filter mask
#As you can see, this makes for a very complicated line of code and it is easy to make syntax errors.
#However, it still works
#- if you are not using a notebook write print(combined_df[ (combined_df['Major'] == 'Art') & (combined_df['Age'] > 30) ])

combined_df[ (combined_df['Major'] == 'Art') & (combined_df['Age'] > 30) ]
```

```python
#Repeating the task above, but with a filter mask 
filter_mask = (combined_df['Major'] == 'Art') & (combined_df['Age'] > 30)
#- if you are not using a notebook write print(combined_df[filter_mask])

combined_df[filter_mask]
```

```python
#Note we can also utilize the .loc method
#- if you are not using a notebook write print(combined_df.loc[filter_mask])

combined_df.loc[filter_mask]
```

```python
#OR
#- if you are not using a notebook write print(combined_df.loc[(combined_df['Major'] == 'Art') & (combined_df['Age'] > 30)])

combined_df.loc[(combined_df['Major'] == 'Art') & (combined_df['Age'] > 30)]
```
