```python
#use your own api key
key = 'insert your api key here'

#api end-point
url = "https://api.nytimes.com/svc/books/v3/reviews.json?author=Stephen+King&api-key=" + key

#making the request
r = requests.get(url)

#if the request is successful then extract the data from the json code that is returned
if(r.status_code == 200):
  data = r.json()
  
  #using json.dumps() to format the json as a string to make it more readable. 
  print(json.dumps(data, sort_keys=True, indent=10))
```

```python
#Extracting the book titles and publication dates from the json and inserting them into two separate lists

book_titles = []
publication_dates = []

#the results key has the data we need
for i in range(0, len(data['results'])):
  print(data['results'][i]['book_title'] + ' ' + data['results'][i]['publication_dt'])
  book_titles.append(data['results'][i]['book_title'])
  publication_dates.append(data['results'][i]['publication_dt'])

```

```python
#creating a dictionary which will later be converted to a dataframe
books = { 
          'title': book_titles,
          'publication dates': publication_dates
        }

```

```python
#Creating the dataframe
import pandas as pd

books_df = pd.DataFrame(books)

books_df.head()
```
