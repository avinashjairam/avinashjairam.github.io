```python
#USING THE NY TIMES API TO FIND THE TOP STORIES
import requests
import json
```

```python
#Utilize your own api key
key = 'INSERT KEY HERE'

#setting which category of stories you wish to access. "home" by default gives you the stories on the homepage of the NY TIMES
# "sports" for example will give you all the top stories about sports  
section = "home"

# f-strings is a new concept in python 3.6. Please read more here - https://docs.python.org/3/whatsnew/3.6.html
url = f"https://api.nytimes.com/svc/topstories/v2/{section}.json?api-key={key}"

#making the api call
r = requests.get(url)

#if the request is successful, store the returned json code in data 
if r.status_code == 200:
  data = r.json()

  #using json.dumps() to print the json as a string and to easier understand the structure 
  print(json.dumps(data, sort_keys=True, indent= 10))
```

```python
#Extracting the url from the json
for i in range(0, len(data['results'])):
  print(data['results'][i]['url'])
```
