```python
#importing pandas
import pandas as pd
```

```python
#What is a series? A series is a python dictionary that has a list as its value.

#A dataframe is a collection of series. That is , a dictionary of series

students = {'Name' : ['Bob', 'Alice', 'Joe', 'Mike', 'Lisa', 'Tom'],
            'Major': ['Accounting', 'Finance', 'Chemistry', 'Art', 'Engineering', 'Art'],
            'GPA'  : [3.5, 3.8, 3.2, 3.7, 4.0, 3.5],
            'Age'  : [22,23,31,40,25,30]
          }

#How do you create a dataframe? 

df = pd.DataFrame(students)

#if you are not using a notebook, use print(df) to print the dataframe

df
```

```python
#.head() - this returns the first 5 rows in the dataset 


#if you are not using a notebook, use print(df.head()) to print the dataframe

df.head()
```

```python
#.head(n) - this returns the first 'n' rows in the dataset

#if you are not using a notebook, use print(df.head(1)) to print the dataframe

df.head(1)
```

```python
#.tail() - this returns the last 5 rows in the dataset

#if you are not using a notebook, use print(df.tail()) to print the dataframe

df.tail()
```

```python
#.tail(n) - this returns the last 'n' rows in the dataset

#if you are not using a notebook, use print(df.tail(3)) to print the dataframe


df.tail(3)
```

```python
#.shape - this returns a tuple of the number of rows and columns in the dataframe

#if you are not using a notebook, use print(df.shape) to print the dataframe

df.shape
```

```python
#.describe() - this returns a numerical breakdown of the statistics of the dataset

#if you are not using a notebook, use print(df.describe()) to print the dataframe

df.describe()
```

```python
#printing the original dataset
#if you are not using a notebook, use print(df) to print the dataframe

df
```

```python
# .T returns a tranpose of the dataset. That is, the indexes (row numbers) and column names are switched

#if you are not using a notebook, use print(df.T) to print the dataframe

df.T
```

```python
#sorting the dataframe

#.sort_index(axis = 1, ascending = False) - sorts the dataset: 0 is used to specify rows, 1 is used to specify columns

#if you are not using a notebook, use print(df.sort_index(axis = 1, ascending = False)) to print the dataframe

df.sort_index(axis = 1, ascending = True)
```

```python
#sorting the dataframe

#.sort_index(axis = 1, ascending = False) - sorts the dataset: 0 is used to specify rows, 1 is used to specify columns

#if you are not using a notebook, use print(df.sort_index(axis = 1, ascending = False)) to print the dataframe

df.sort_index(axis = 0, ascending = False)
```

```python
#sorting the dataframe by column 

#.sort_values(by = 'name of column') - This returns a dataframe sorted by a particular column in ascending order

#if you are not using a notebook, use print(df.sort_values(by = 'GPA')) to print the dataframe


df.sort_values(by = 'GPA')
```

```python
#sorting the dataframe by column 

#.sort_values(by = 'name of column' , ascending = False) - This returns a dataframe sorted by a particular column in descending order

#if you are not using a notebook, use print(df.sort_values(by = 'GPA')) to print the dataframe


df.sort_values(by = 'GPA', ascending = False)
```

```python
#.sort_values(by = 'name of column' , ascending = Trye) - This returns a dataframe sorted by a particular column in ascending order

#if you are not using a notebook, use print(df.sort_values(by = 'Age')) to print the dataframe

df.sort_values(by = 'Age', ascending=True)

```

```python
#sorting the dataframe by column 

#.sort_values(by = 'name of column' , ascending = False) - This returns a dataframe sorted by a particular column in descending order

#if you are not using a notebook, use print(df.sort_values(by = 'Age')) to print the dataframe


df.sort_values(by = 'Age', ascending = False)
```
