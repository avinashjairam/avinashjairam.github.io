```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
url = 'https://en.wikipedia.org/wiki/List_of_United_States_cities_by_population'

page = requests.get(url)
```

```python
print(page.content)
```

```python
soup = BeautifulSoup(page.content, 'html.parser')

print(soup.prettify())
```

```python
tables = soup.find_all('table', class_ = 'sortable')

correct_table = tables[0]

#checking if this is the correct table

#print(correct_table.find_all('th'))


rank = []
city = []
state = []
census_2020 = []
census_2010 = []
pct_change = []
land_area_sqm = []
land_area_sqk = []
pop_density_sqm = []
pop_density_sqk = []
location = []

rows = correct_table.find_all('tr')

for row in rows:
  cells = row.find_all('td')

  if(len(cells) == 11):
    
    rank.append(cells[0].get_text().rstrip('\n'))
    city.append(cells[1].get_text().rstrip('\n'))
    state.append(cells[2].get_text().rstrip('\n'))
    census_2020.append(cells[3].get_text().rstrip('\n'))
    census_2010.append(cells[4].get_text().rstrip('\n'))
    pct_change.append(cells[5].get_text().rstrip('\n'))
    land_area_sqm.append(cells[6].get_text().rstrip('\n'))
    land_area_sqk.append(cells[7].get_text().rstrip('\n'))
    pop_density_sqm.append(cells[8].get_text().rstrip('\n'))
    pop_density_sqk.append(cells[9].get_text().rstrip('\n'))
    location.append(cells[10].get_text().rstrip('\n'))
```

```python
df = pd.DataFrame()
df['rank'] = rank
df['city'] = city
df['state'] = state
df['census_2020'] = census_2020
df['census_2010'] = census_2010
df['pct_change'] = pct_change
df['land_area_sqm'] = land_area_sqm
df['land_area_sqk'] = land_area_sqk
df['pop_density_sqm'] = pop_density_sqm
df['pop_density_sqk'] = pop_density_sqk
df['location'] = location


```
