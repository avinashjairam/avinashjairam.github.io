```python
#Creating a basic line plot 

#import matplotlib library

from matplotlib import pyplot as plt 
```

```python
# x - axis values
x = [5, 2, 9, 4, 7]

# y - axis values

y = [10, 5, 8, 4 ,2]
```

```python
#creating the plot

plt.plot(x, y)
plt.title("Random Line Graph")
plt.xlabel("x - values")
plt.ylabel("y - values")

#function to show or display the plot
plt.show()
```
